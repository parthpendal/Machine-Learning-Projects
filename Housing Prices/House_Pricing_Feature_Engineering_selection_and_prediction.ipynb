Advanced Housing Prices- Feature Engineering
The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead

Dataset to downloaded from the below link
https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

We will be performing all the below steps in Feature Engineering

Missing values
Temporal variables
Categorical variables: remove rare labels
Standarise the values of the variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.concat([pd.read_csv('train.csv'), pd.read_csv('test.csv')])
dataset.head()
```

## Missing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing

Features_nan = [Feature for Feature in dataset.columns if dataset[Feature].isnull().sum()>0 and dataset[Feature].dtype=='O' ]

for feature in Features_nan:
  print((dataset[feature].isnull().sum())*100/dataset[feature].isna().count(),"% null values for feature  ", feature)
```

```python
## Replace missing value with a new label
def replace_nan_with_missing(dataset, Features_nan):
  data=dataset.copy()
  for feature in Features_nan:
    data[feature]=data[feature].replace(to_replace =np.nan,value="Missing")
  return data

dataset=replace_nan_with_missing(dataset,Features_nan)



```

```python
dataset.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_features_with_nan=[feature for feature in dataset.columns if dataset[feature].dtype!='O' and dataset[feature].isna().sum()>0]
numerical_features_with_nan
```

```python
## Replacing the numerical Missing Values
for feature in numerical_features_with_nan:
  if feature!='SalePrice':
    median =dataset[feature].median()
    dataset[feature]=dataset[feature].replace(to_replace=np.nan,value=median )

dataset[numerical_features_with_nan].head(5)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]

dataset.head(20)
```

```python
#Function to use get dummies for all the columns

categorical_features = [Feature for Feature in dataset.columns if  dataset[Feature].dtype=='O' ]

def category_onehot_multcols(multcolumns,df):
    for fields in multcolumns:
        df1=pd.get_dummies(df[fields],drop_first=True)
        df.drop([fields],axis=1,inplace=True)
        df=pd.concat([df,df1],axis=1)
    return df

dataset=category_onehot_multcols(categorical_features,dataset)
dataset.head()

```

```python
# Removing duplicate columns
dataset=dataset.loc[:,~dataset.columns.duplicated()]
```

## Feature Scaling

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
# Segregating Train and test data. Record with salesprice as null will be our test data
df_test=data[data.SalePrice.isna() ]
df_test.head()
```

```python
# Segregating Train and test data. Record with salesprice as not null will be our train data
df_train=data[data.SalePrice.notnull() ]
df_train.head()
```

```python
# creating x_train and y_train
x_train=df_train.drop(["SalePrice","Id"],axis=1)
y_train=df_train[["SalePrice"]]
```
