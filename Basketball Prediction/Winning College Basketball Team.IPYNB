```python
import matplotlib.pyplot as plt
```

```python
def get_scatter_map(x_axis,y_axis,df,x_axis_name,y_axis_name):
    Analysis_df=df[[x_axis,y_axis,"POSTSEASON"]]
    colors = ["Red","Yellow","Blue","Black","Brown","Purple","Green","Orange","Pink"]
    color_ind=0
    plt.figure(figsize=(10,10))
    for elimination in Analysis_df["POSTSEASON"].unique():
        elimination_df=Analysis_df[Analysis_df["POSTSEASON"]==elimination]
        plt.scatter(x_axis,y_axis,data=elimination_df, c=colors[color_ind],label=elimination)
        color_ind+=1
        plt.legend()
    plt.xlabel(x_axis_name)
    plt.ylabel(y_axis_name)
    plt.show()
get_scatter_map("ORB","DRB",df,"Offensive Rebound Rate","Offensive Rebound Rate Allowed")
get_scatter_map("TOR", "TORD",df,'Turnover Rate',"steam Rate")
get_scatter_map("ADJOE","ADJDE",df,'Adjusted Offensive Efficiency','Adjusted Defence Efficiency')
get_scatter_map("FTR","FTRD",df,"Free Throw Rate","Free Throw Rate Allowed")
get_scatter_map("2P_O","2P_D",df,"Two-Point Shooting Percentage","Two-Point Shooting Percentage Allowed")
get_scatter_map("3P_O","3P_D",df,"Three-Point Shooting Percentage","Three-Point Shooting Percentage Allowed")
```

# Generating KNN model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=df.dropna()
dataframe=df[["G","W","ADJOE","ADJDE","BARTHAG","EFG_O","EFG_D","TOR","TORD","ORB","DRB","FTR","FTRD","2P_O","2P_D","3P_O","3P_D","ADJ_T","WAB"]]
std_df=StandardScaler().fit_transform(dataframe)
x_train,x_test,y_train,y_test=train_test_split(std_df,df["POSTSEASON"],test_size=0.2, random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh=KNeighborsClassifier(n_neighbors=5)
neigh.fit(x_train,y_train)
```

```python
y_hat=neigh.predict(x_test)
```

```python
print(y_test[:5])
print(y_hat[:5])
```

# Determine the accuracy for the first 15 values of k the on the validation data

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test[:15], y_hat[:15]))
```

# Determine the minimum value for the parameter that improves results on validation data.

```python
acc_scr=[]
for k in range(1,100):
    neigh=KNeighborsClassifier(n_neighbors=k)
    neigh.fit(x_train,y_train)
    y_hat=neigh.predict(x_test)
    acc_scr.append(metrics.accuracy_score(y_test, y_hat))
```

```python
plt.plot(range(1,100),acc_scr)
```
